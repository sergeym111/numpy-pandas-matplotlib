```python
import pandas as pd
```

```python
import numpy as np
```

```python
dict = {
    'author_id': [1, 2, 3],
    'author_name': ['Тургенев', 'Чехов', 'Островский']
}
```

```python
dict
```

```python
authors = pd.DataFrame(dict)
```

```python
authors 
```

```python
dict_1 = {
    'author_id': [1, 1 ,1, 2, 2, 3, 3],
    'book_title': ['Отцы и дети', 
                   'Рудин', 
                   'Дворянское гнездо', 
                   'Толстый и тонкий', 
                   'Дама с собачкой',
                   'Гроза',
                   'Таланты и поклонники',
                  ],
    'price': [450, 300, 350, 500, 450, 370, 290]
}
```

```python
book = pd.DataFrame(dict_1)
book
```

# 002_задание

```python
authors_price = pd.merge(authors, book, on='author_id', how='outer')
authors_price
```

# 003_задание

```python
top5 = authors_price.sort_values(by='price').tail(5)
top5
```

```python
top5 = authors_price.nlargest(5, 'price')
top5
```
